```python
from geopy.geocoders import Nominatim

def get_area_name(latitude, longitude):
    geolocator = Nominatim(user_agent="area_finder")
    location = geolocator.reverse((latitude, longitude), exactly_one=True)
    address = location.raw['address']
    
    # Try to get the city first
    area = address.get('city', '')
    if not area:
        # If city is not available, try to get the town
        area = address.get('town', '')
        if not area:
            # If town is not available, try to get the village
            area = address.get('village', '')
            if not area:
                # If village is not available, try to get the state
                area = address.get('state', '')
                if not area:
                    # If state is not available, try to get the administrative division (e.g., county)
                    area = address.get('county', '')
    return area

# Assuming Fire is your dataset containing latitude and longitude columns
new_data_frame = pd.DataFrame(columns=['Latitude', 'Longitude', 'Location'])

for index, row in Fire.iterrows():
    latitude = row['latitude']
    longitude = row['longitude']
    state = get_area_name(latitude, longitude)
    Fire["State"]= state
    print(f"Latitude: {latitude}, Longitude: {longitude}, State: {state}")
```
