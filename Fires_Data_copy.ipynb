```python
# Query All Records in the fire_archive Table
fire_archive_df = pd.read_sql_query("SELECT * from fire_archive", con)
print(fire_archive_df.head())
```

```python
# Query All Records in the fire_nrt Table
fire_nrt_df = pd.read_sql_query("SELECT * from fire_nrt", con)
print(fire_nrt_df.head())
```
